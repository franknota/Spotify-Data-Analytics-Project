```python
from IPython.display import display

# Assuming 'data_df' is your DataFrame
display(data_df.head())  # Displays the first 5 rows of the DataFrame

```

```python
# Calculate the most played artist
most_played_artist = data_df['artistName'].value_counts().idxmax()
most_played_artist_count = data_df['artistName'].value_counts().max()

# Calculate the most played track
most_played_track = data_df['trackName'].value_counts().idxmax()
most_played_track_count = data_df['trackName'].value_counts().max()

print(f"The most played artist is {most_played_artist} with {most_played_artist_count} plays.")
print(f"The most played track is {most_played_track} with {most_played_track_count} plays.")
```

```python
# Set the number of top elements to display
top_n = 10

# Get the top N artists
top_artists = data_df['artistName'].value_counts().head(top_n)
top_tracks = data_df['trackName'].value_counts().head(top_n)

# Set up the matplotlib figure
plt.figure(figsize=(14, 8))

# First subplot for top artists
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
sns.barplot(y=top_artists.index, x=top_artists.values, palette='viridis')
plt.title('Top {} Most Played Artists'.format(top_n))
plt.xlabel('Play Count')
plt.ylabel('Artist')

# Second subplot for top tracks
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
sns.barplot(y=top_tracks.index, x=top_tracks.values, palette='magma')
plt.title('Top {} Most Played Tracks'.format(top_n))
plt.xlabel('Play Count')
plt.ylabel('Track')

# Display the plots
plt.tight_layout()
plt.show()
```
